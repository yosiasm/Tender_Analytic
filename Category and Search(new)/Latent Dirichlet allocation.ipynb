```python
kata= 'Pengadaan Living Furniture span class='
kata.rsplit('<')[0]
```

```python
df = pd.DataFrame({'preprocess':[kata.rsplit('<')[0].lower() for kata in df]}) #menghilangkan script html dan lowercasing
```

```python
import re
df['preprocess'] = df['preprocess'].map(lambda x: re.sub('[,\.!?]', '', x)) #menghilangkan punctiation
df
```

```python
df['preprocess'] = df['preprocess'].str.replace('\d+', '') #menghilangkan data tahun
```

```python
stop = ['dan','tahun','untuk']
df['preprocess'] = df['preprocess'].apply(lambda x: ' '.join([item for item in x.split() if item not in stop])) #menghilangkan stopwords
df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid')
def plot_top_10(count_data, count_vectorizer):
    import matplotlib.pyplot as plt
    words = count_vectorizer.get_feature_names()
    total_counts = np.zeros(len(words))
    for t in count_data:
        total_counts+=t.toarray()[0]
    
    count_dict = (zip(words, total_counts))
    count_dict = sorted(count_dict, key=lambda x:x[1], reverse=True)[0:10]
    words = [w[0] for w in count_dict]
    counts = [w[1] for w in count_dict]
    x_pos = np.arange(len(words)) 
    
    plt.figure(2, figsize=(15, 15/1.6180))
    plt.subplot(title='10 kata yg sering muncul')
    sns.set_context("notebook", font_scale=1.25, rc={"lines.linewidth": 2.5})
    sns.barplot(x_pos, counts, palette='husl')
    plt.xticks(x_pos, words, rotation=90) 
    plt.xlabel('words')
    plt.ylabel('counts')
    plt.show()

# membuat count vectorizer dan menampilkan 10 kata yg sering muncul

count_vectorizer = CountVectorizer(stop_words='english')
count_data = count_vectorizer.fit_transform(df['preprocess'])
plot_top_10(count_data, count_vectorizer)
```

```python
#membuat model lda
from sklearn.decomposition import LatentDirichletAllocation as LDA
 
#menampilkan hasil
def print_topics(model, count_vectorizer, n_top_words):
    words = count_vectorizer.get_feature_names()
    for topic_idx, topic in enumerate(model.components_):
        print("\nKategori #%d:" % topic_idx)
        print(" ".join([words[i]
                        for i in topic.argsort()[:-n_top_words - 1:-1]]))
        
# tunning jumlah topic dan kata
number_topics = 10
number_words = 7
lda = LDA(n_components=number_topics, n_jobs=-1)
lda.fit(count_data)
# Print hasil
print("Kategori dengan LDA:")
print_topics(lda, count_vectorizer, number_words)
```

```python
#membuat fungsi untuk mengkategori berdasarkan model LDA diatas
words = count_vectorizer.get_feature_names()
def predict_topic(nama_tender):
    nama_tender=nama_tender.lower()
    n_top_words = 10
    list_bobot = []
    for topic_idx, topic in enumerate(lda.components_):
        keyword = [words[i] for i in topic.argsort()[:-n_top_words - 1:-1]]
        #print(topic_idx,topic,keyword)
        bobot=0
        for weight,key in zip(topic,keyword):
            if key in nama_tender:
                
                bobot = bobot + weight
                list_bobot.append(bobot)
                #print(key,bobot)
    if list_bobot == []:
        return topic_idx+1
    else:
        return [i for i, j in enumerate(list_bobot) if j == max(list_bobot)][0]

print(predict_topic('kean'))
```

```python
nama_tender = "Pengadaan Perangkat Pengolah Data dan Komunikasi "
kategori_tender = predict_topic(nama_tender)
kategori_tender
```
