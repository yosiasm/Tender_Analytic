```python
import dash
import dash_table
import dash_core_components as dcc
import dash_html_components as html
import plotly.graph_objs as go
import numpy as np
import pandas as pd
```

```python
app=dash.Dash(__name__)
```

```python
df = pd.read_csv('data sirup update.csv')
df=df.drop(columns=['Unnamed: 0'])
tabel_jumlah_tender_tahun2020=df[['nama','waktu']].loc[df['waktu'] != '2021-01-01'].loc[df['waktu'] != '2023-10-01'].loc[df['waktu'] != '2023-04-01'].groupby(['waktu'],as_index=False).count()
tabel_jumlah_tender_tahun2020.rename(columns = {'nama':'Jumlah Tender'}, inplace = True)
tabel_jumlah_tender_tahun2020.head()
#df.head()
```

```python
tabel1=pd.read_csv('tabel instansi_hps_kategori.csv')
tabel1[['Jasa Konsultansi Badan Usaha', 'Jasa Konsultansi Perorangan','Pekerjaan Konstruksi', 'Pengadaan Barang', 'Jasa Lainnya']].astype(int)
tabel1.columns
```

```python
tabel2=pd.read_csv('tabel instansi_metode.csv')
tabel2.head()
```

```python
df.loc[df['waktu'] == '2020-01-01']
```

```python
app.layout = html.Div([
    
    dcc.Graph(id='chart-hps',
              figure={
                  'data':[{'x': tabel1.instansi,'y':tabel1.hps_mean, 'type' : 'bar', 'name':'rerata-hps'}],
                  'layout':{'title':'Perbandingan Instansi dengan HPS',
                            'yaxis':{'title': 'Rerata HPS'},
                            'legendText':'True'
                            #'xaxis':{'title': 'Nama Instansi'},
                           }
              }
    ),
      dcc.Graph(id='chart-jumlah-tender',
              figure={
                  'data':[{'x': tabel1.instansi,'y':tabel1.jumlah_tender, 'type' : 'bar', 'name':'jumlah-tender'},
                         {'x': tabel1.instansi,'y':tabel1['Jasa Konsultansi Badan Usaha'], 'type' : 'bar', 'name':'jasa-konsultansi-badan-usaha'},
                         {'x': tabel1.instansi,'y':tabel1['Jasa Konsultansi Perorangan'], 'type' : 'bar', 'name':'jasa-konsultansi-perorangan'},
                         {'x': tabel1.instansi,'y':tabel1['Pekerjaan Konstruksi'], 'type' : 'bar', 'name':'pekerjaan-konstruksi'},
                         {'x': tabel1.instansi,'y':tabel1['Pengadaan Barang'], 'type' : 'bar', 'name':'pengadaa-barang'},
                         {'x': tabel1.instansi,'y':tabel1['Jasa Lainnya'], 'type' : 'bar', 'name':'jasa-lainnya'}],
                  'layout':{'title':'Perbandingan Instansi dengan Jumlah Tender Berdasarkan Kategori'}
              }),
     html.H1(children='TABLE PERBANDINGAN INSTANSI DENGAN METODE TENDER',
            style={'textAlign':'center'}),
    
     dash_table.DataTable(id='table-hps',
                          columns=[{"name": i, "id": i} for i in tabel2.columns],
                          data=tabel2.to_dict('records'), style_table={'overflowX': 'scroll','overflowY': 'scroll', 'maxHeight': '300px',},
                          sort_action='native', selected_columns=[], selected_rows=[],
                          filter_action="native", sort_mode="multi", column_selectable="single",
                         ),
    
    html.H2(children='TABLE PERBANDINGAN INSTANSI DENGAN HPS DAN KATEGORI TENDER',
            style={'textAlign':'center'}),
    
     dash_table.DataTable(id='table-metode',
                          columns=[{"name": i, "id": i} for i in tabel1.columns],
                          data=tabel1.to_dict('records'), style_table={'overflowX': 'scroll','overflowY': 'scroll', 'maxHeight': '300px',},
                          sort_action='native', selected_columns=[], selected_rows=[],
                          filter_action="native", sort_mode="multi", column_selectable="single",
                         ),
    
    dcc.Graph(id='chart-2020',
             figure={'data':[go.Scatter(x=tabel_jumlah_tender_tahun2020['waktu'],
                                       y=tabel_jumlah_tender_tahun2020['Jumlah Tender'])
                            ],
                    'layout':go.Layout(title='JUMLAH TENDER 2020',
                                      xaxis={'title':'Bulan Tahun'},
                                      yaxis={'title':'Jumlah Tender'})}),
    
    
    html.H3(children='TABLE TENDER TAHUN 2020',
           style={'textAlign':'center'}),
    
    
    dash_table.DataTable(id='table',
                         columns=[{"name": i, "id": i} for i in df.columns],
                         data=df.to_dict('records'), page_size=8, style_table={'overflowX': 'scroll'},
                         sort_action='native', selected_columns=[], selected_rows=[],
                         filter_action="native", sort_mode="multi", column_selectable="single",
                        )
])   
```

```python
if __name__=='__main__':
    app.run_server(port=4050)
```
