```python
df3=df1[['pagu','6','7','8', 'hps', 'hps_detail', 'penawaran_pemenang']]
df3.head()
```

```python
df3.rename(columns = {'8':'kategori','6':'metode pemilihan','7':'metode evaluasi'}, inplace = True)
```

```python
df3=df3.dropna()
df3
```

```python
hps_number=[]
for i in df3['hps']:
    if "M" in i:
        print(i)
        i=i.replace(" M", "").replace(',','.')
        new_value=float(i)*1000000000
        print(new_value)
        hps_number.append(new_value)
    elif "Jt" in i:
        print(i)
        i=i.replace(" Jt", "").replace(',','.')
        new_value=float(i)*1000000
        print(new_value)
        hps_number.append(new_value)
```

```python
hps_detail_number=[]
for i in df3['hps_detail']:
    if "Rp " in i:
        print(i)
        for a in range (1,99):
            i=i.replace("Rp ", "").replace(',00','').replace(','+str(a),'').replace('.','').replace(',0'+str(a),'')
        new_value=float(i)
        print(new_value)
        hps_detail_number.append(new_value)
```

```python
pagu_number=[]
for i in df3['pagu']:
    if "Rp " in i:
        print(i)
        i=i.replace("Rp ", "").replace(',00','').replace('.','').replace(',26','')
        new_value=float(i)
        print(new_value)
        pagu_number.append(new_value)
```

```python
penawaran_pemenang_number=[]
for i in df3['penawaran_pemenang']:
    if "Rp " in i:
        print(i)
        for a in range (1,99):
            i=i.replace("Rp ", "").replace(',00','').replace(','+str(a),'').replace('.','').replace(',0'+str(a),'')
        new_value=float(i)
        print(new_value)
        penawaran_pemenang_number.append(new_value)
```

```python
df3['pagu_number'] = pagu_number
df3['hps_number'] = hps_number
df3['hps_detail_number'] = hps_detail_number
df3['penawaran_pemenang_number']= penawaran_pemenang_number
df3
```

```python
# buat variabel y
y = modeled_dataset['penawaran_pemenang_number']

# buat variabel X
X = modeled_dataset.iloc[:,1:2]
# train-test split
X_train, X_validation, y_train, y_validation = train_test_split(X,y, test_size=0.2, random_state=1000)
# buat objek LinearRegression
linreg = LinearRegression()
reg = linear_model.Lasso(alpha=0.1)
# buat model
reg.fit(X_train, y_train)
```

```python
#X_train
```

```python
# lihat nilai R^2
reg.score(X_train, y_train)
```

```python
# membuat prediksi
hasil_prediksi = reg.predict(X_validation)
#hasil_prediksi = np.exp(hasil_prediksi) + 1
# lihat hasil prediksi
hasil_prediksi
```

```python
# melihat hasil root mean squared error
mean_squared_error(y_validation, hasil_prediksi)
```
